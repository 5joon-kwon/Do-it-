```python
from sklearn.datasets import load_diabetes  # 당뇨병 환자 데이터
```

```python
diabetes = load_diabetes()
print(type(diabetes))
print(diabetes.keys()) # 다운받은 dataset은 data(input), target(label)이 존재
```

```python
# input(data), label(target) 확인
print(diabetes.data.shape, diabetes.target.shape) # tuple로 출력 : 원소 하나면 (4,)로 출력됨
# input : 442x10
# label : 442
```
